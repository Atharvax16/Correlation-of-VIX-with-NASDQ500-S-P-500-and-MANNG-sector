```python
data_vix['observation_date'] = pd.to_datetime(data_vix['observation_date'])
data_nasdq['observation_date'] = pd.to_datetime(data_nasdq['observation_date'])
```

```python
data_vix = data_vix.sort_values('observation_date')
data_nasdaq = data_nasdq.sort_values('observation_date')
```

```python
data_merged = pd.merge(
    data_vix,
    data_nasdq,
    on='observation_date',
    how='inner',
    suffixes=('_VIX', '_NASDAQ')
)
```

```python
data_merged.head()
```

```python
data_merged.info()
```

```python
missing_merged = data_merged[data_merged['NASDAQ100'].isna()]
missing_merged
```

```python
data_clean = data_merged.dropna(subset=['VIXCLS', 'NASDAQ100']).copy()
data_clean.info()
```

SAME NO OF ROWS

```python
data = data_merged.copy()
```

```python
pearson_lvl = data['VIXCLS'].corr(data['NASDAQ100'], method='pearson')
spearman_lvl = data['VIXCLS'].corr(data['NASDAQ100'], method='spearman')
print(pearson_lvl, spearman_lvl)
```

```python
data['nasdaq_ret'] = np.log(data['NASDAQ100'] / data['NASDAQ100'].shift(1))
data = data.dropna(subset=['nasdaq_ret'])

pearson_ret = data['VIXCLS'].corr(data['nasdaq_ret'], method='pearson')
spearman_ret = data['VIXCLS'].corr(data['nasdaq_ret'], method='spearman')
print(pearson_ret, spearman_ret)
```

```python
plt.figure(figsize=(10,4))
plt.plot(data['observation_date'], data['NASDAQ100'], label='NASDAQ100')
plt.ylabel('Index Level')
plt.legend()
plt.title('NASDAQ100 over time')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(data['observation_date'], data['VIXCLS'], color='orange', label='VIX')
plt.ylabel('VIX Level')
plt.legend()
plt.title('VIX over time')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10,4))

ax1.plot(data['observation_date'], data['nasdaq_ret'],
         color='tab:blue', label='Nasdaq log return')
ax1.set_ylabel('Nasdaq log return', color='tab:blue')

ax2 = ax1.twinx()
ax2.plot(data['observation_date'], data['VIXCLS'],
         color='tab:orange', alpha=0.6, label='VIX')
ax2.set_ylabel('VIX level', color='tab:orange')

plt.title('Nasdaq daily log returns and VIX over time')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10,4))

ax1.plot(data['observation_date'], data['NASDAQ100'],
         color='tab:blue', label='Nasdaq  return')
ax1.set_ylabel('Nasdaq  return', color='tab:blue')

ax2 = ax1.twinx()
ax2.plot(data['observation_date'], data['VIXCLS'],
         color='tab:orange', alpha=0.6, label='VIX')
ax2.set_ylabel('VIX level', color='tab:orange')

plt.title('Nasdaq daily returns and VIX over time')
plt.show()
```

FOR BASIC UNDERSTANDING

VIX --> market fear/uncertainty index for the US stock market

High VIX → market expects big moves (often in a crisis, lots of fear).

Low VIX → market expects calm, smaller moves.

NASDAQ100 PRICE TRENDS UP WHILE VIX SPENDS MOST OF THE TIME LOW WITH SHORT SPIKES

ON BIG DOWN DAYS FOR NASDAQ, VIX IS ELEVATED AT THE SAME TIME.

WHEN MARKET GETS SCARED WHICH MEANS VIX HIGH,STOCK PRICES OFTEN FALLING

WHEN MARKET IS CALM WHICH MEANS VIX LOW, STOCK PRICES ARE HIGH

THERE IS A INVERSE CORRELATION/NEGATIVE CORRELATION IN BETWEEN THE VIX AND NASDQ100.
